```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("stock_file.csv")
```

```python
dataset.shape
```

```python
dataset.head
```

```python
dataset.info()
```

```python
# drop tuples with NA as these data are usually not related to IPO
dataset.dropna()
```

```python
datasest = dataset.dropna()
```

```python
# drop 認購不足 data for simplicity
dataset.drop(dataset[dataset['over_sub_ratio'] == '認購不足'].index, inplace=True)
```

```python
dataset.dropna(axis=0, how="any", inplace=True)
```

```python
dataset.info()
```

```python
dataset['over_sub_ratio'] = dataset['over_sub_ratio'].apply(pd.to_numeric, errors='coerce')
```

```python
dataset.info()
```

```python
# filling "null" items with 0 for "over_sub_ratio"  
dataset['over_sub_ratio'] = dataset['over_sub_ratio'].fillna(0)
```

```python
dataset.info()
```

```python
dataset['final_offer_price'] = dataset['final_offer_price'].apply(pd.to_numeric, errors='coerce')
```

```python
dataset['upper_offer_pr'] = dataset['upper_offer_pr'].apply(pd.to_numeric, errors='coerce')
```

```python
dataset['lower_offer_pr'] = dataset['lower_offer_pr'].apply(pd.to_numeric, errors='coerce')
```

```python
dataset.info()
```

```python
# check the null value for offer price columns
dataset[dataset.isnull().any(axis=1)]
```

```python
# only one column, hard code the correct value
dataset['lower_offer_pr'] = dataset['lower_offer_pr'].fillna(1399)
dataset['upper_offer_pr'] = dataset['upper_offer_pr'].fillna(1399)
dataset['final_offer_price'] = dataset['final_offer_price'].fillna(1190)
```

```python
dataset.info()
```

```python
# remove % sign from the last two columns and change the columns to numeric
dataset['allot_odds'] = dataset['allot_odds'].str.replace('%', '')
dataset['firstday_return'] = dataset['firstday_return'].str.replace('%', '')
```

```python
dataset['allot_odds'] = dataset['allot_odds'].apply(pd.to_numeric, errors='coerce')
dataset['firstday_return'] = dataset['firstday_return'].apply(pd.to_numeric, errors='coerce')
```

```python
# reduce by 100 for last two columns
dataset['allot_odds'] = dataset['allot_odds'] / 100
dataset['firstday_return'] = dataset['firstday_return'] / 100
```

```python
dataset.info()
```

```python
# cleaning market_cap column
dataset['market_cap'] = dataset['market_cap'].str.replace(',', '')
```

```python
dataset
```

```python
# expand the market cap range into two new columns, "mkt_cap_low" and "mkt_cap_high"
dataset[['mkt_cap_low', 'mkt_cap_high']] = dataset['market_cap'].str.split('-', expand=True)
```

```python
dataset.info()
```

```python
# check the null value for any 'unsplit' columns
dataset[dataset.isnull().any(axis=1)]
```

```python
# convert the data from object to numeric
dataset['mkt_cap_high'] = dataset['mkt_cap_high'].apply(pd.to_numeric, errors='coerce')
```

```python
dataset['mkt_cap_low'] = dataset['mkt_cap_low'].apply(pd.to_numeric, errors='coerce')
```

```python
# check for NaN values for further cleaning
dataset[dataset.isnull().any(axis=1)]
```

```python
# filling NaN value with market cap (in this case as 
# there is not market cap range, mkt_cap_high is equal mkt_cap_low = market_cap)
dataset['mkt_cap_high'] = dataset['mkt_cap_low'].fillna(dataset['market_cap'])
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
# no null data exists, final check for datatype
dataset.info()
```

```python
# Finally, rearrange 'firstday_return' to be the last column for learning as 'Y'
column_titles = ['ticker', 'list_date', 'market_cap', 'mkt_cap_low', 'mkt_cap_high', 'lower_offer_pr', 'upper_offer_pr', 'final_offer_price', 'over_sub_ratio', 'allot_odds', 'firstday_return']
dataset = dataset.reindex(columns = column_titles)
```
